# Reconocimiento de patrones usando ORB

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Función para mostrar imágenes en una cuadrícula con Matplotlib

```python
def mostrar_imagenes(filas, columnas, titulos, imagenes, w, h, imagen_gris, format=False, v=False):
    """
    Muestra una lista de imágenes en una cuadrícula.

    Parámetros:
      filas, columnas: dimensiones de la cuadrícula.
      titulos: lista de títulos para cada imagen.
      imagenes: lista de imágenes (en formato BGR o escala de grises).
      w, h: ancho y alto de la figura.
      imagen_gris: tupla/lista booleana indicando si cada imagen es en escala de grises.
      format: si True, se configuran límites en los ejes.
      v: si True, se invierte el orden de las filas de la imagen (efecto vertical).
    """
    fig, ax = plt.subplots(filas, columnas, figsize=(w, h))
    if filas * columnas == 1:
        ax = [ax]
    else:
        ax = ax.ravel() if hasattr(ax, 'ravel') else [ax]
    
    for i, a in enumerate(ax):
        if i < len(imagenes):
            img = imagenes[i]
            if v:
                img = img[::-1, :]  # Invierte las filas (volteo vertical)
            a.set_title(titulos[i], fontsize=10)
            if format:
                a.set(xlim=(0, 20), ylim=(0, 20))
            # Si la imagen es a color se convierte de BGR a RGB para visualizar correctamente
            if not imagen_gris[i]:
                if img.ndim == 3:
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                a.imshow(img)
            else:
                a.imshow(img, cmap='gray')
            a.axis('off')
        else:
            a.axis('off')
    plt.tight_layout()
    plt.show()
```

## Carga de imágenes

```python
# Actualiza las rutas a las imágenes según tu entorno
path1 = '../data/formulario.png'
path2 = '../data/formularioRelleno-original.png'
path3 = '../data/formularioRelleno-rotado.png'

# Comprobación de existencia de los ficheros
assert os.path.exists(path1), f"El fichero {path1} no existe"
assert os.path.exists(path2), f"El fichero {path2} no existe"
assert os.path.exists(path3), f"El fichero {path3} no existe"

# Lectura de imágenes con OpenCV
formulario_vacio = cv2.imread(path1)
formulario_relleno_original = cv2.imread(path2)
formulario_relleno_rotado = cv2.imread(path3)

# Mostrar imágenes originales en una cuadrícula 1x3
titulos = ['Formulario', 'Relleno', 'Relleno Rotado']
imagenes = [formulario_vacio, formulario_relleno_original, formulario_relleno_rotado]
mostrar_imagenes(1, 3, titulos, imagenes, w=20, h=14, imagen_gris=(False, False, False))
```

## Detección y emparejamiento de puntos de interés con ORB

```python
# Crear objeto ORB
orb = cv2.ORB_create()

# Detectar puntos clave y calcular descriptores en cada imagen
kp1, des1 = orb.detectAndCompute(formulario_vacio, None)
kp2, des2 = orb.detectAndCompute(formulario_relleno_original, None)
kp3, des3 = orb.detectAndCompute(formulario_relleno_rotado, None)

# Crear objeto BFMatcher usando la métrica Hamming (adecuada para ORB)
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Emparejar descriptores entre la imagen del formulario y las otras imágenes
coincidencias1 = bf.match(des1, des2)
coincidencias2 = bf.match(des1, des3)

# Ordenar los matches según la distancia (los de menor distancia son mejores)
coincidencias1 = sorted(coincidencias1, key=lambda x: x.distance)
coincidencias2 = sorted(coincidencias2, key=lambda x: x.distance)

# Seleccionar los 100 mejores matches
mejores_coincidencias1 = coincidencias1[:100]
mejores_coincidencias2 = coincidencias2[:100]
```

## Cálculo de las matrices de transformación (Homografía)

```python
# Obtener las coordenadas de los puntos de interés para cada conjunto de matches

# Para la imagen Relleno Original:
src_pts1 = np.float32([kp1[m.queryIdx].pt for m in mejores_coincidencias1]).reshape(-1, 1, 2)
dst_pts1 = np.float32([kp2[m.trainIdx].pt for m in mejores_coincidencias1]).reshape(-1, 1, 2)

# Para la imagen Relleno Rotado:
src_pts2 = np.float32([kp1[m.queryIdx].pt for m in mejores_coincidencias2]).reshape(-1, 1, 2)
dst_pts2 = np.float32([kp3[m.trainIdx].pt for m in mejores_coincidencias2]).reshape(-1, 1, 2)

# Estimar matrices de homografía usando RANSAC
M1, mask1 = cv2.findHomography(dst_pts1, src_pts1, cv2.RANSAC, 5.0)
M2, mask2 = cv2.findHomography(dst_pts2, src_pts2, cv2.RANSAC, 5.0)
```

## Aplicar matrices de transformación (Warp Perspective) con fondo blanco

```python
# Se utiliza borderValue=(255,255,255) para rellenar con fondo blanco.
ancho = formulario_vacio.shape[1]
alto = formulario_vacio.shape[0]

formulario_relleno_original_trans = cv2.warpPerspective(
    formulario_relleno_original, M1, (ancho, alto), borderValue=(255, 255, 255)
)
formulario_relleno_rotado_trans = cv2.warpPerspective(
    formulario_relleno_rotado, M2, (ancho, alto), borderValue=(255, 255, 255)
)
```

## Dibujar coincidencias y guardar resultados

```python
# Dibujar coincidencias entre formulario vacío y formulario rellenado original
img_matches1 = cv2.drawMatches(formulario_vacio, kp1, 
                               formulario_relleno_original, kp2, 
                               mejores_coincidencias1, None, flags=2)
# Dibujar coincidencias entre formulario vacío y formulario rellenado rotado
img_matches2 = cv2.drawMatches(formulario_vacio, kp1, 
                               formulario_relleno_rotado, kp3, 
                               mejores_coincidencias2, None, flags=2)

# Guardar las imágenes con coincidencias en disco
cv2.imwrite('resultado_coincidencias_original.png', img_matches1)
cv2.imwrite('resultado_coincidencias_rotado.png', img_matches2)

# Guardar también las imágenes transformadas
cv2.imwrite('formulario_relleno_original_transformado.png', formulario_relleno_original_trans)
cv2.imwrite('formulario_relleno_rotado_transformado.png', formulario_relleno_rotado_trans)

# Mostrar las imágenes con coincidencias en una cuadrícula 1x2
mostrar_imagenes(1, 2, 
                 ['', ''], 
                 [img_matches1, img_matches2], 
                 w=20, h=14, imagen_gris=(False, False))
```

## Mostrar imágenes transformadas con fondo blanco

```python
titulos_trans = ['', '', '']
imagenes_trans = [formulario_vacio, formulario_relleno_original_trans, formulario_relleno_rotado_trans]

mostrar_imagenes(1, 3, titulos_trans, imagenes_trans, w=20, h=14, imagen_gris=(False, False, False))
```
