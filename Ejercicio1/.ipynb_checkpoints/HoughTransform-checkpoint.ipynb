## Aplicación de la Transformada de Hough

```python
# Importa las libreías necesarias
import cv2 as cv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# obtiene los parámetros de las líneas obtenidas con los algoritmos Hough de openCV
def parametros(lineas, lineasP):
    rho_media_H = np.mean(lineas, 0)
    rho_media_HP = np.mean(lineasP, 0)
    m = -(rho_media_HP[0, 3] - rho_media_HP[0, 1]) / (rho_media_HP[0, 2] - rho_media_HP[0, 0])
    # Obtiene la media de las pendientes de ambas rectas en radianes 
    pendiente_H = rho_media_H[0, 1]
    pendiente_HP = -np.arctan(-1 / m)
    return pendiente_H, pendiente_HP
```

```python
# Función para dibujar las líneas en la imagen a partir de los 
# parámetros obtenidos en cv.HoughLines()
def dibuja_lineas(lineas, imagen):
    for linea in lineas:
        rho, theta = linea[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * a)
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * a)
        cv.line(imagen, (x1, y1), (x2, y2), (0, 0, 255), 2)
```

```python
# Función para dibujar las líneas en la imagen a partir de los 
# parámetros obtenidos en cv.HoughLinesP()
def dibuja_lineas_P(lineas, imagen):
    for i in range(0, len(lineas)):
        l = lineas[i][0]
        cv.line(imagen, (l[0], l[1]), (l[2], l[3]), (0,0,255), 1, cv.LINE_AA)
```

```python
# Localización de la imagen
img_path = '../data/textoMolinos.png'
```

```python
# Lee la imagen original
original_BGR = cv.imread(img_path, cv.IMREAD_COLOR)
original_RGB = cv.cvtColor(original_BGR, cv.COLOR_BGR2RGB)
```

```python
# Pasa la imagen a escala de grises
escala_grises = cv.cvtColor(original_RGB, cv.COLOR_BGR2GRAY)

# Obtiene los bordes a partir de la imagen en escala de grises
bordes = cv.Canny(escala_grises, 50, 150, apertureSize=3)
bordesP = np.copy(bordes)

# Calcula las lineas usando el algoritmo Hough y el Hough mejorado
lineas_H = cv.HoughLines(bordes, 1, np.pi / 180, 250)
lineas_HP = cv.HoughLinesP(bordesP, 1, np.pi / 180, 50, None, 50, 10)

# Crea dos imágenes auxiliares para dibujar las líneas
original_con_lineas_H = np.copy(original_RGB)
original_con_lineas_HP = np.copy(original_RGB)

# Dibuja las líneas en las imágenes auxiliares
dibuja_lineas(lineas_H, original_con_lineas_H)
dibuja_lineas_P(lineas_HP, original_con_lineas_HP)

# Establece los títulos para la representación de las imágenes
titulos = ['Obtención de bordes con Canny',
           'Lineas detectadas co HoughLines',
           'Lineas detectadas con HoughLinesP', '']

# Establece las imágenes para los ejes
imagenes = [bordes, original_con_lineas_H, original_con_lineas_HP, None]

# Muestra las imágenes
src.mostrar_imagenes(1, 3, titulos, imagenes, 20, 14, (True, False, False))
```

```python
# Obtiene los parámetros necesarios para la matriz de rotación
rows, cols, ch = original_RGB.shape
pendiente_H, pendiente_HP = parametros(lineas_H, lineas_HP)

print(f'pendiente H: {pendiente_H}, pendiente HP: {pendiente_HP}')
# obtiene la matriz de rotación a partir de las lineas obtenidas
# con los algoritmos cv.HoughLines y cv.HoughLinesP
M = cv.getRotationMatrix2D(
    ((cols - 1) / 2.0, (rows - 1) / 2.0), 
    np.rad2deg(pendiente_H - np.pi / 2),
    1)

MP = cv.getRotationMatrix2D(
    ((cols - 1) / 2.0, (rows - 1) / 2.0),
    np.rad2deg(pendiente_HP - np.pi / 2), 
    1)

# Realiza la transformación Afín a partir de las 
# matrices obtenidas
transformacion_H = cv.warpAffine(original_RGB, M, (cols, rows),    
                                 flags=cv.INTER_LINEAR,
                                 borderMode=cv.BORDER_CONSTANT,
                                 borderValue=(255, 255, 255))
transformacion_HP = cv.warpAffine(original_RGB, MP, (cols, rows),    
                                  flags=cv.INTER_LINEAR,
                                  borderMode=cv.BORDER_CONSTANT,
                                  borderValue=(255, 255, 255))

# Establece los títulos para la representación de las imágenes
titulos = ['Imagen Original',
           'Transformación a partir de HoughLines',
           'Transformación a partir de HoughLinesP', '']

# Establece las imágenes para los ejes
imagenes = [original_RGB, transformacion_H, transformacion_HP, None]

# Muestra las imágenes
src.mostrar_imagenes(1, 3, titulos, imagenes, 20, 14, (True, False, False))
```

```python
# Obtiene la imagen transformada en escala de grises
transformacion_HP_gris = cv.cvtColor(transformacion_HP, cv.COLOR_BGR2GRAY)
```

```python
# Variables globales para almacenar los valores actuales de bloque y c
bloque_mean = 3  # Valor inicial para bloque
c_mean = 2       # Valor inicial para c

# Función para aplicar el umbral adaptativo
def umbral_adaptativo_mean(bloque, c):
    global bloque_mean, c_mean
    bloque_mean= bloque  # Actualizar el valor de bloque
    c_mean = c            # Actualizar el valor de c
    
    # Aplicar el umbral adaptativo
    imagen_umbral = cv.adaptiveThreshold(transformacion_HP_gris,
                                         255,
                                         cv.ADAPTIVE_THRESH_MEAN_C,
                                         cv.THRESH_BINARY_INV, 
                                         bloque, 
                                         c)
    
    # Mostrar la imagen usando la función personalizada
    src.mostrar_imagenes(1, 1, ['Umbralización adaptativo media'], 
                         [imagen_umbral], 20, 14, (True, ))

# Crear los sliders interactivos
interact(umbral_adaptativo_mean, 
         bloque=widgets.IntSlider(min=3, max=30, step=2, value=5), 
         c=widgets.IntSlider(min=0, max=50, step=1, value=30));
```

```python
# Variables globales para almacenar los valores actuales de bloque_gauss y c_gauss
bloque_gauss = 3  # Valor inicial para bloque_gauss
c_gauss = 2       # Valor inicial para c_gauss

# Función para aplicar el umbral adaptativo gaussiano
def umbral_adaptativo_gauss(bloque, c):
    global bloque_gauss, c_gauss
    bloque_gauss = bloque  # Actualizar el valor de bloque_gauss
    c_gauss = c            # Actualizar el valor de c_gauss
    
    # Aplicar el umbral adaptativo gaussiano
    imagen_umbral = cv.adaptiveThreshold(transformacion_HP_gris,
                                         255,
                                         cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                                         cv.THRESH_BINARY_INV, 
                                         bloque, 
                                         c)
    
    # Mostrar la imagen usando la función personalizada
    src.mostrar_imagenes(1, 1, ['Umbralización adaptativo gausiano'], 
                         [imagen_umbral], 20, 14, (True, ))

# Crear los sliders interactivos
interact(umbral_adaptativo_gauss, 
         bloque=widgets.IntSlider(min=3, max=30, step=2, value=3), 
         c=widgets.IntSlider(min=0, max=50, step=1, value=2));
```

```python
# Obtiene los contornos de las letras.
# Otras maneras alternativas de obtener los contornos de las imágenes
def contornos_config(config):
    if config == 1:
        #print(f'opción {config}: cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE')
        return cv.findContours(umbral, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    elif config == 2:
        #print(f'opción {config}: cv.RETR_TREE, cv.CHAIN_APPROX_NONE')
        return cv.findContours(umbral, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
    elif config == 3:
        #print(f'opción {config}: cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE')
        return cv.findContours(umbral, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    elif config == 4:
        #print(f'opción {config}: cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE')
        return cv.findContours(umbral, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
```

```python
# Dibujar contornos
contornos_img = np.zeros((umbral.shape[0], umbral.shape[1], 3), dtype=np.uint8)

config = 3

contours, hierarchy = contornos_config(config)
for i in range(len(contours)):
    cv.drawContours(contornos_img, contours, i, (0,255,0), 1, cv.LINE_8, hierarchy, 0)
    
src.mostrar_imagenes(1, 1, ['Representación de los contornos'], 
                     [contornos_img], 20, 14, (True, ))
```

```python
# Estira la dimensión más estrecha
def estirar(char):
    width, height = char.shape
    relacion_aspecto = width / height
    if width > height:
        width = 20
        height = int(width / relacion_aspecto)
    else:
        height = 20
        width = int(height * relacion_aspecto)
    return width, height
```

```python
# Normaliza el carcacter manteniendo 
# la relación de aspecto 
def mantener_aspecto(width, height):
    if width > height:
        width_ = 20
        height_ = int(height * (20 / width))
    else:
        height_ = 20
        width_ = int(width * (20 / height))
    return width_, height_
```

```python
chars = []
centrados = []
estirados = []
conservados = []
for contour in contours:
    # Obtiene el rectángulo a partir de cada contorno
    x, y, w, h = cv.boundingRect(contour)

    # Obtiene el caracter original y lo almacena
    char = umbral[y:y+h, x:x+w]
    chars.append(char)

    # Obtiene las nuevas dimensiones para estirar el caracter
    width, height = estirar(char)
  
    # Cambia el tamaño y lo almacena
    estirado = cv.resize(char, (width, height))
    estirados.append(estirado)

    # Obtiene las nuevas dimensiones conservando la relación de aspecto
    width, height = mantener_aspecto(w, h)
  
    # Cambia el tamaño y lo almacena
    conservado = cv.resize(char, (width, height))
    conservados.append(conservado)

    # Crea una imagen vacía de 20x20
    centrado = np.zeros((20, 20), dtype=np.uint8)

    # Calcula la posición en la que debe colocarse el caracter normalizado
    start_x = int((20 - width) / 2)
    start_y = int((20 - height) / 2)

    # Coloca el caracter normalizado en el centro de la imagen y lo almacena
    centrado[start_y:start_y+height, start_x:start_x+width] = conservado
    centrados.append(centrado)
```

```python
def letras(i): 
    # Muestra las imágenes
    src.mostrar_imagenes(1, 4, ['Caracter original', 'Estirado', 'Aspecto conservado', 'Centrado'], 
                     [chars[i], estirados[i],conservados[i], centrados[i]], 
                     6, 6, (True, True, True, True), True, True)
interact(letras, i=widgets.IntSlider(min=0, max=len(chars), step=1, value=719));
```

```python
def detectar (chars, area_min=36, area_max=192):    
    #aspect_ratio_min, aspect_ratio_max = 0.3, 2.5
    correctos = []
    solapados = []
    errores = []
    for i, char in enumerate(chars):
        h, w = char.shape
        area = h * w
        if area > area_max:
            solapados.append(char)
        elif area < area_min:
            errores.append(char)
        else:
            correctos.append(char)
    
    return correctos, solapados, errores

correctos, solapados, errores = detectar(chars)
```

```python

# Texto de entrada
text = ("¿Qué vio Cervantes en los molinos para convertirlos en gigantes? "
        "«Ves allí, amigo Sancho Panza, donde se descubren treinta o poco más "
        "desaforados gigantes con quien pienso hacer batalla. ¿Qué gigantes?, "
        "dijo Sancho Panza. Aquellos que allí ves, respondió su amo, de los brazos largos, "
        "que los suelen tener algunos de casi dos leguas». Este es uno de los extractos más "
        "famosos de El Quijote, junto con “En un lugar de La Mancha…” o «La libertad, Sancho, "
        "es uno de los más preciosos dones…”, entre otros mucho más. La célebre historia de "
        "los molinos convertidos en gigantes procede de la elección de esos aparatos, por "
        "parte del gran Miguel de Cervantes, debido a que fueron las máquinas más modernas e "
        "impactantes de la época, con gigantescas dimensiones: doce metros de altura y aspas "
        "de ocho metros y medio. En la actualidad, solo nueve molinos -del medio millar que "
        "pobló La Mancha hace 400 años- conservan su maquinaria de fabricación: tres en Campo "
        "de Criptana (Ciudad Real), uno en Mota del Cuervo (Cuenca), dos en Consuegra, uno en "
        "Madridejos, uno en El Romeral y otro en Camuñas, todas las localidades pertenecientes "
        "a la provincia de Toledo. Igualmente, todos se ponen en funcionamiento para realizar "
        "moliendas al menos una vez al año, pues, como asiente Vicente Casero, “cuando quedan "
        "un año o dos sin actividad, luego necesitan tareas de rehabilitación y de puesta a punto”.")

# --------------------------------------------
# 5. Cálculo de métricas básicas
# --------------------------------------------
numero_caracteres_reales = len(text)
numero_chars = len(chars)
numero_correctos = len(correctos)
numero_solapados = len(solapados)
numero_errores = len(errores)

# --------------------------------------------
# 6. Cálculo de porcentajes
# --------------------------------------------
porcentaje_acierto = (numero_correctos / numero_chars) * 100
porcentaje_solapados = (numero_solapados / numero_chars) * 100
porcentaje_errores = (numero_errores / numero_chars) * 100

# --------------------------------------------
# 7. Construir DataFrame con TODAS las métricas
# --------------------------------------------
data = {
    "Aspecto": [
        "Número de caracteres reales",
        "Número total de contornos (chars)",
        "Número de caracteres correctos",
        "Número de caracteres solapados (criterio área)",
        "Número de caracteres con errores (ruido)",
        "Porcentaje de acierto",
        "Porcentaje de solapados",
        "Porcentaje de errores"
    ],
    "Resultados": [
        numero_caracteres_reales,
        numero_chars,
        numero_correctos,
        numero_solapados,
        numero_errores,
        f"{porcentaje_acierto:.2f} %",
        f"{porcentaje_solapados:.2f} %",
        f"{porcentaje_errores:.2f} %"
    ]
}

df_resultados = pd.DataFrame(data)

# --------------------------------------------
# 8. Mostrar DataFrame y exportar a CSV
# --------------------------------------------
display(df_resultados)
df_resultados.to_csv("resultados_segmentacion.csv", index=False)
print("Tabla de resultados exportada a 'resultados_segmentacion.csv'.")
```
